CREATING BAR GRAPHS TO SHOW COUNTS

```python
def rank_performance(stock_price):
    if (stock_price <=10):
        return("poor")
    elif(stock_price <=50):
        return("satisfactory")
    else:
        return("stellar")
```

```python
bb["Close"].apply(rank_performance) #apply the fuction to close column
```

```python
bb["Close"].apply(rank_performance).value_counts() #count the number of times 
                                          #each value shows.       
```

```python
plt.style.use("ggplot")
bb["Close"].apply(rank_performance).value_counts().plot() #we will get only a line chart
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind="bar")# default
                                       # is kind="line".
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind="barh") # reverse
#barh=bar horizontal                                              #the axis.
```

```python
df = pd.DataFrame.from_dict(
    {
        'Year': [2016, 2017, 2018, 2019, 2020, 2021],
        'Men': [300, 320, 310, 305, 325, 330],
        'Women': [350, 325, 300, 305, 380, 355],
        'Total': [650,645, 610, 610, 705, 685]
    }
)
```

```python
df
```

```python
plt.bar(x=df['Year'],height=df['Total'],color='yellow')
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.show()
```

```python
col=['yellow','yellow','yellow','yellow','red','yellow']
plt.bar(x=df['Year'],height=df['Total'],color=col)
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.show()
```

```python
col=['red' if (bar == max(df['Total'])) else 'grey' for bar in df['Total']]
plt.bar(x=df['Year'],height=df['Total'],color=col)
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)                 # USING LIST COMPREHENSION
plt.ylabel('Sales',fontsize=12)
plt.show()
```

```python
col=['yellow','yellow','yellow','yellow','red','yellow']
plt.barh(y=df['Year'],width=df['Total'],color=col)
plt.title('Time over Sales',fontsize=18)
plt.xlabel('Sales',fontsize=12)
plt.ylabel('Year',fontsize=12)
plt.show()
```

```python
col=['yellow','yellow','yellow','yellow','red','yellow']
plt.bar(x=df['Year'],height=df['Total'],color=col,width=0.4)
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)              #TO CHANGE WIDTH OF BAR
plt.ylabel('Sales',fontsize=12)
plt.show()
```

```python
width=0.4
plt.bar(x=df['Year'],height=df['Men'],color='blue',width=0.4,label='Men')
plt.bar(x=df['Year']+width,height=df['Women'],color='orange',width=0.4,label='Women') 
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.bar(x=df['Year'],height=df['Men'],color='blue',label='Men')
plt.bar(x=df['Year'],height=df['Women'],color='orange',label='Women',bottom=df['Men']) 
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend(loc='lower right')
plt.show()
```

```python
df['men proportion']=(df['Men']/df['Total'])*100
df['women proportion']=(df['Women']/df['Total'])*100
plt.bar(x=df['Year'],height=df['men proportion'],color='blue',label='Men')
plt.bar(x=df['Year'],height=df['women proportion'],color='orange',label='Women',bottom=df['men proportion']) 
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend(loc='lower right')
plt.show()
```

```python
plt.style.use('ggplot')
df['men proportion']=(df['Men']/df['Total'])*100
df['women proportion']=(df['Women']/df['Total'])*100
plt.bar(x=df['Year'],height=df['men proportion'],label='Men')
plt.bar(x=df['Year'],height=df['women proportion'],label='Women',bottom=df['men proportion']) 
plt.title('Sales over Time',fontsize=18)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend(loc='lower right')
plt.show()
```
