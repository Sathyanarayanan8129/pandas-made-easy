# WEB SCRAPPING USING PANDAS

```python
#sending request
url='https://www.bankbazaar.com/reviews.html'
page=requests.get(url) #we will receive an html code
page.text   #html code as a text
```

NOW WE ARE GOING TO PARSE HTML CODE WITH BEAUTIFULSOUP4

```python
soup=BeautifulSoup(page.text,'html.parser')
soup
```

NOW TO READ DATA

```python
rev_text_elm=soup.find_all(class_="text_here review-desc-more") 
```

```python
print(rev_text_elm)
```

```python
review_text=[]
for item in rev_text_elm:
    review_text.append(item.text) # the text in rev_text_ elm will be appended
                                   # will be appended to review_text in a loop                                       
review_text
```

```python
user_name=[]
user_name_elm=soup.find_all(class_="reviewer-profile")
for item in user_name_elm:
    user_name.append(item.text)
    
user_name
```

```python
# read bank name  also right click and click inspect to access each attribute
#to access image, use div to avoid redundancy
bank_name=[]
bank_name_elm=soup.find_all('div',{'class':'review-bank-title'}) #find all division whose class attribute is review bank title
bank_name_elm
```

```python
for item in bank_name_elm:
    bank_name.append(item.find('img')['alt'])
bank_name    
```

```python
# now create a dataframe by combining all
final_array=[]
for text,user,bank in zip(review_text,user_name,bank_name):
    final_array.append({'Review':text,'User':user,'Bank':bank})
df=pd.DataFrame(final_array)
df
```

# another example of web scrap

```python
url='https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'
link=requests.get(url).text
link
```

```python
soup=BeautifulSoup(link,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
soup.find_all("a")
```

```python
# to get only the links ,we iterate over all a tags with get(href)
all_link=soup.find_all("a")
for link in all_link:
    print(link.get('href'))
```

```python
all_tables=soup.find_all('table')
print(all_tables)
```

```python
right_table=soup.find('table',class_="wikitable sortable")
print(right_table)
```

```python
table_links=right_table.find_all('a')
table_links
```

```python
country=[]
for link in table_links:
    country.append(link.get('title'))
print(country)
```

```python
df=pd.DataFrame(country)
df=df.rename({0:'Country'},axis=1)
```
