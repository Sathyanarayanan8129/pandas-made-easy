```python
import pandas as pd
```

```python
df=pd.read_csv('http://bit.ly/kaggletrain')
df
```

```python
df['Sex_no']=df['Sex'].map({'male': 1,'female':0})
```

```python
df.loc[0:4,'Sex':'Sex_no']
```

```python
df['Name_Length']=df['Name'].apply(len)
```

```python
df.loc[0:2,'Name':'Name_Length']
```

```python
import numpy as np
```

```python
#another example using numpy function
df['Fare_ceil']=df['Fare'].apply(np.ceil)
```

```python
df.loc[0:3,['Fare','Fare_ceil']] 
```

```python
df['Name'].str.split(',').str.get(0).head()
```

```python
df['Name'].str.split(',',expand=True)
```
